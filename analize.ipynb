```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
path = "action_history_20251108_135516.csv"
df = pd.read_csv(path)
columns = ["timestamp", "joint_L_1", "joint_L_2", "joint_R_1", "joint_R_2"]
for col in columns:
    if col == "timestamp":
        continue
    df[col] = df[col] * (180.0 / np.pi)  # rad -> deg
df = df[columns]
df.head()
```

```python
col = "joint_R_2"
plt.figure(figsize=(20, 6))
plt.scatter(df['timestamp'], df[col], label=col, s=2)
plt.xlabel('Timestamp')
plt.ylabel(col + ' Position')
plt.legend()
plt.grid()
plt.show()
```
